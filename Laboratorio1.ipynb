```python
from google.colab import drive
import pandas as pd
import chardet

# Montar Google Drive
drive.mount('/content/drive', force_remount=True)

# Rutas del archivos CSV
ruta_archivo = "/content/drive/MyDrive/Finance_Trends_enriquecido.csv"


# Función para detectar la codificación
def detectar_codificacion(ruta):
    with open(ruta, 'rb') as f:
        result = chardet.detect(f.read())
    encoding = result['encoding']
    print(f" {ruta.split('/')[-1]} → Codificación detectada: {encoding}")
    return encoding

# Función para leer CSV con detección automática
def leer_csv(ruta):
    try:
        encoding = detectar_codificacion(ruta)

        df = pd.read_csv(ruta, sep= None, engine= "python", encoding=encoding)
        # sep=None con engine='python' → pandas detecta automáticamente si el separador es , o ;.
        print(f" {ruta.split('/')[-1]} cargado correctamente. ({df.shape[0]} filas, {df.shape[1]} columnas)")
        return df
    except FileNotFoundError:
        print(f" Error: No se encontró el archivo en {ruta}")
        return None
    except Exception as e:
        print(f" Error inesperado al leer {ruta.split('/')[-1]}: {e}")
        return None

#  Leer archivos
df_Finanzas = leer_csv(ruta_archivo)


```

```python
# Al Observar que también los registros estan en otro idioma prcedemos a pasarlos al idioma español
# para una mejor compresion de los datos
from googletrans import Translator
import time
import pandas as pd

# Inicializar el traductor
translator = Translator()

# Definir las columnas a excluir de la traducción
columnas_excluir = ['Genero'] # Excluir la columna 'Genero' de la traducción automática

# Obtener las columnas de tipo 'object' a traducir
object_columns = df_Cambio_Idioma.select_dtypes(include='category').columns.tolist()
columnas_a_traducir = [col for col in object_columns if col not in columnas_excluir]

print(f"Columnas a traducir: {columnas_a_traducir}")

# Diccionario para almacenar los mapeos de traducción por columna
mapeos_traduccion = {}

# Traducir valores únicos para cada columna
for columna in columnas_a_traducir:
    valores_unicos = df_Cambio_Idioma[columna].unique()
    print(f"\nTraduciendo valores únicos para la columna: '{columna}'...")

    mapeo_columna = {}
    for valor in valores_unicos:
        # Asegurarse de que el valor no sea nulo y sea una cadena de texto
        if pd.notna(valor) and isinstance(valor, str):
            try:
                # Intentar detectar el idioma de origen y traducir a español
                # source_lang = translator.detect(valor).lang # Opcional: detectar idioma
                translated_text = translator.translate(valor, dest='es').text
                mapeo_columna[valor] = translated_text
                # print(f"  '{valor}' -> '{translated_text}'") # Opcional: imprimir traducciones
                time.sleep(0.1) # Pequeña pausa para evitar sobrecargar el servicio
            except Exception as e:
                print(f"  Error al traducir '{valor}': {e}")
                mapeo_columna[valor] = valor # Mantener el valor original en caso de error
        else:
             mapeo_columna[valor] = valor # Mantener valores no string/nulos


    mapeos_traduccion[columna] = mapeo_columna
    print(f"Traducción de columna '{columna}' completada.")

# Aplicar los mapeos de traducción al DataFrame
print("\nAplicando traducciones al DataFrame...")
for columna, mapeo in mapeos_traduccion.items():
    df_Cambio_Idioma[columna] = df_Cambio_Idioma[columna].map(mapeo)
    print(f"Traducciones aplicadas a la columna '{columna}'.")


print("\n Traducción de columnas de texto completada.")
print("\nVista previa del DataFrame con columnas traducidas:")
display(df_Cambio_Idioma.head())

#  Ruta de salida para guardar el archivo traducido
ruta_salida_traducido = "/content/drive/MyDrive/Finance_Trends_traducido1.csv"

# Guardar el DataFrame traducido como CSV
df_Cambio_Idioma.to_csv(ruta_salida_traducido, index=False, encoding="utf-8-sig")

print(f"\n Archivo traducido guardado correctamente en '{ruta_salida_traducido}'")
```
